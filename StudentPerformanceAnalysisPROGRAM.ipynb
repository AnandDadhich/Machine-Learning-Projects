```python
#Set a Student Passing marks for each three subject's individually
passmarks=40
```

```python
df.columns
```

```python
#Creating a New column pass_math, this column will tell us that the student is pass or fail
df["pass_math"]=np.where(df["math score"]>passmarks,"Pass","Fail")

plt.figure(figsize=(5,5),dpi=300)
df["pass_math"].value_counts().plot.pie()
plt.title("Pass/Fail in Maths")
plt.xlabel("Status")
plt.ylabel("Count")
```

```python
df['pass_math'].value_counts()
```

```python
df.columns
```

```python
#Creating a new column pass_read, this column will tell us that the student is Pass or fail
df['pass_read']=np.where(df['reading score']>passmarks,"Pass","Fail")

plt.figure(figsize=(5,5),dpi=300)
df['pass_read'].value_counts().plot.pie(colors=["lightblue","lightgreen"])
plt.title("Pass/Fail in Reading")
plt.xlabel("Count")
plt.ylabel("Status")
```

```python
#Creating a new column pass_writing, this column will tell us that the student is Pass or Fail
df['pass_writing']=np.where(df['writing score']>passmarks,"Pass","Fail")

plt.figure(figsize=(5,5),dpi=300)
plt.title("Pass/Fail in Writing")
df['pass_writing'].value_counts().plot.pie(colors=['lightblue','darkgreen'])
plt.xlabel("Count")
plt.ylabel("Status")
```
