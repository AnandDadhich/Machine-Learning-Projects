```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data=make_blobs(n_samples=600,n_features=2,centers=3,cluster_std=1,random_state=42)
```

```python
data[0]   #actual data
```

```python
data[1]   #classes for reference purpose only
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap="rainbow")
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=3).fit(data[0])
```

```python
model.cluster_centers_
```

```python
#model.labels_
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=model.labels_,cmap="rainbow")
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap="rainbow")
plt.title("Original")
plt.subplot(1,2,2)
plt.scatter(data[0][:,0],data[0][:,1],c=model.labels_,cmap="rainbow")
plt.title("Kmeans Clustering")
```
